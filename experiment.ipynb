PROJECT : NEXT WORD PREDICTION USING LSTM

```python
## DATA COLLECTION
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd
```

```python
## LOAD THE DATASET
data = gutenberg.raw('shakespeare-hamlet.txt')
```

```python
## SAVE DATA TO A FILE
with open('hamlet.txt','w') as file:
    file.write(data)
```

```python
## DATA PREPROCESSING

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
## LOAD THE DATASET

with open('hamlet.txt','r') as file:
    text = file.read().lower()

## TOKENIZE THE TEXT  -- CREATING INDEX FOR EACH WORD

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
```

```python
total_words
```

```python
tokenizer.word_index
```

```python
## CREATE INPUT SEQUENCE

input_sequences = []
for line in text.split('\n'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1,len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
## PAD SEQUENCES
max_seq_len = max(len(x) for x in input_sequences)
```

```python
max_seq_len
```

```python
input_sequences = np.array(pad_sequences(input_sequences,maxlen=max_seq_len,padding='pre'))
```

```python
input_sequences
```

```python
## CREATING PREDICITORS AND LABEL
import tensorflow as tf
x,y = input_sequences[:,:-1], input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
## CONVERT Y INTO CATEGORIES

y = tf.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
y
```

```python
## SPLIT THE DATA INTO TRAIN AND TEST
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

## DEFINE THE MODEL
model = Sequential()
model.add(Embedding(total_words, 100, input_length = max_seq_len-1))
model.add(LSTM(150, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words, activation = "softmax"))

# This line forces the model to build itself
# The None is for the batch size, which can be flexible
model.build((None, max_seq_len))

## COMPILE THE MODEL
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
## TRAIN THE MODEL

history = model.fit(x_train,y_train,epochs=220, validation_data=(x_test,y_test),verbose=1)
```

```python
## FUNCTION TO PREDICT THE NEXT WORD
def predict_next_word(model, tokenizer, text, max_seq_len):
    token_list = tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_seq_len:
        token_list = token_list[-(max_seq_len-1):]
    token_list = pad_sequences([token_list],maxlen = max_seq_len-1, padding='pre')
    predicted = model.predict(token_list, verbose = 0)
    predicted_word_index = np.argmax(predicted, axis = 1)
    for word,index in tokenizer.word_index.items():
        if index == predicted_word_index:
            return word
    return None
```

```python
input_text = "To be or not to be"
print(f"Input Text : {input_text}")
max_seq_len = model.input_shape[1]+1
next_word = predict_next_word(model,tokenizer,input_text, max_seq_len)
print(f"Next Word Predixtion : {next_word}")
```
